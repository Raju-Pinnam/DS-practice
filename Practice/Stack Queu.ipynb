# STACK

stack - creates new stack empty 

push(item)

pop()

peek - returns top element but not removes

isEmpty - return stack is empty or not

size - returns number of items 

```python
class Stack:
    
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return len(self.items) == 0
    
    def push(self, item):
        self.items.append(item)
    
    def pop(self):
        if self.is_empty():
            raise IndexError("pop from empty stack")
        return self.items.pop()
    
    def peek(self):
        if self.is_empty():
            raise IndexError("peek from empty stack")
        return self.items[-1]
    
    def size(self):
        return len(self.items)
```

# Queue

```python
class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.insert(0, item)
    
    def dequeue(self):
        if self.is_empty():
            raise IndexError("dequeue from empty queue")
        return self.items.pop()
    
    def size(self):
        return len(self.items)
```

# Deque

```python
class Deque:
    
    def __init__(self):
        self.items = []
    
    def frontpush(self, item):
        self.items.insert(0, item)
    
    def rearpush(self, item):
        self.items.append(item)
    
    def frontpop(self):
        if self.is_empty():
            raise IndexError("frontpop from empty deque")
        return self.items.pop(0)
    
    def pop(self):
        self.items.pop()
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)
```

```python
def balance_check(s):
    
    if len(s) % 2 != 0:
        return False
    
    opening = set('({[')
    matches = set([('(', ')'), ('[', ']'), ('{', '}')])
    
    stack = []
    
    for paren in s:
        if paren in opening:
            stack.append(paren)
        else:
            if len(stack) == 0:
                return False
            last_open = stack.pop()
            if (last_open, paren) not in matches:
                return False
    
```
