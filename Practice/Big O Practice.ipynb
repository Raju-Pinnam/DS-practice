```python
def sum1(n):
    final_sum = 0
    for i in range(1, n + 1):
        final_sum += i
    return final_sum
def sum2(n):
    return n * (n + 1) / 2
```

```python
sum1(10), sum2(10)
```

```python
%timeit sum1(100)
```

```python
%timeit sum2(100)
```

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

n = np.linspace(0, 10, 100)
labels = ['Constant', 'Logarithamic', 'Linear', 'Log Linear', 'Quadratic', 'Cubic', 'Exponential']
big_o = [np.ones(n.shape), np.log(n), n, n*np.log(n), n**2, n**3, 2**n]

plt.figure(figsize=(12, 10))
plt.ylim(0, 50)

for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])
```
