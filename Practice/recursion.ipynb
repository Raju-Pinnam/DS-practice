```python
def sum_mod(n):
    div, mod = divmod(n, 10)
    if div == 0:
        return mod
    return mod + sum_mod(div)
```

```python
sum_mod(1)
```

```python
def reverse_string(s):
    if len(s) == 0:
        return s
    return s[-1] + reverse_string(s[:-1])
```

```python
def permute(s):
    out = []
    print(s)
    if len(s) == 1:
        out.append(s[0])
    else:
        # for every letter in str
        for i, let in enumerate(s):
            # for every permutation of the remaining letters
            for perm in permute(s[:i] + s[i+1:]):
                # add the letter to the front of each permutation of the remaining letters
                out.append(let + perm)
    return out
```

```python
permute(s)
```

```python
def feb_iter(n):
    a, b = 0, 1
    st = ''
    for i in range(0, n):
        st += str(a) + ' '
        a, b = b, a + b
    st += str(a)
    print(st)
    return a
```

```python
def rec_coins(target, coins):
    
    min_coins = target
    
    if target in coins:
        return 1
    
    else:
        for i in  [c for c in coins if c <= target]:
            num_coins = 1 + rec_coins(target - i, coins)
            if num_coins < min_coins:
                min_coins = num_coins
    return min_coins
```
