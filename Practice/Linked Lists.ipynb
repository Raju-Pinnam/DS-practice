# Singly Linked List

```python
class Node:
    
    def __init__(self, val):
        self.val = val
        self.next = None
```

```python
a = Node(1)
b = Node(2)
c = Node(3)
d = Node(4)
```

```python
a.next = b
b.next = c
c.next = d
```

# Double Linked List

```python
class DoubleLinkedListNode:
    
    def __init__(self, val):
        self.val = val
        self.next = None
        self.prev = None
```

```python
a = DoubleLinkedListNode(1)
b = DoubleLinkedListNode(2)
c = DoubleLinkedListNode(3)
d = DoubleLinkedListNode(4)
```

```python
a.next = b
b.prev = a
b.next = c
c.prev = b
c.next = d
d.prev = c
```

# Interview Problems

```python
def cycle_check(node):
    marker1 = node
    marker2 = node
    
    while marker2 and marker2.next:
        marker1 = marker1.next
        marker2 = marker2.next.next
        
        if marker1 == marker2:
            return True
    return False
```

Linked List Reversal

```python
def reverse(head):
    current = head
    previous = None
    next = None
    # a -> b -> c -> d
    while current:
        next = current.next     # b
        current.next = previous # None
        previous = current      # a
        current = next      # b
        # b -> a -> None
    return previous
```

N th to Last Node Problem

```python
def nth_to_last_node(n, head):
    left_pointer = head
    right_pointer = head
    
    for i in range(n-1):
        if not right_pointer.next:
            raise Exception("n is larger than the linked list")
        right_pointer = right_pointer.next
    while right_pointer.next:
        left_pointer = left_pointer.next
        right_pointer = right_pointer.next
    return left_pointer
```
